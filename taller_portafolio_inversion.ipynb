## Parte 1 — Tuplas (Información fija)
Las tuplas son útiles cuando los datos **no cambian**.

```python
empresas = ("Apple", "Microsoft", "Amazon", "Tesla", "Google")
print(empresas)
```

### Ejercicio
Imprime:
- La primera empresa
- La tercera empresa
- La última empresa

```python
print(empresas[0])
print(empresas[-3])
print(empresas[-1])
```

## Parte 2 — Lista (Precio de las acciones)

```python
precios_acciones = [185, 320, 145, 250, 275]
print(precios_acciones)
```

### Ejercicio
Imprime:
- Precio de Apple
- Precio de Tesla
- Cantidad de empresas en el portafolio usando len()

```python
print(precios_acciones[0])
print(precios_acciones[-2])
print(len(empresas))
```

## Parte 3 — Diccionario (Cantidad de acciones)

```python
acciones_portafolio = {
    "Apple": 10,
    "Microsoft": 5,
    "Amazon": 8,
    "Tesla": 4,
    "Google": 6
}
print(acciones_portafolio)
```

### Ejercicio
Imprime:
- Acciones de Apple
- Acciones de Google

```python
print(acciones_portafolio["Apple"])
print(acciones_portafolio["Google"])
```

## Parte 4 — Valor invertido

```python
valor_apple = precios_acciones[0] * acciones_portafolio["Apple"]
print(valor_apple)
```

### Ejercicio
Calcula el valor invertido en Microsoft, Amazon, Tesla y Google.

```python
valor_Microsoft = precios_acciones[0] * acciones_portafolio["Microsoft"]
print(valor_Microsoft)
valor_Amazon = precios_acciones[0] * acciones_portafolio["Amazon"]
print(valor_Amazon)
valor_Tesla = precios_acciones[0] * acciones_portafolio["Tesla"]
print(valor_Tesla)
valor_Google = precios_acciones[0] * acciones_portafolio["Google"]
print(valor_Google)
```

## Parte 5 — Lista del portafolio

```python
valor_portafolio = []
valor_portafolio.append(valor_apple)
valor_portafolio.append(valor_Microsoft)
valor_portafolio.append(valor_Amazon)
valor_portafolio.append(valor_Tesla)
valor_portafolio.append(valor_Google)
```

```python
print("Valor total del portafolio:", sum(valor_portafolio))
print("Promedio:", sum(valor_portafolio)/len(valor_portafolio))
print("Número de empresas:", len(valor_portafolio))
```
