# Simulacion basica de inversion

## Vamos a modelar un proyecto de inversion

# Declarar las variables financieras

```python
inversion_inicial = 50000000 #esta variable es de tipo entero
```

```python
tasa_descuento=0.10 #Esta variable es <class 'float'>
```

```python
flujo_1=1500000
flujo_2=1800000
flujo_3=1200000
```

```python
print(type(inversion_inicial))
print(type(tasa_descuento))
print(type(flujo_1))
```

# Operacion basica financiera

```python
total_flujos = flujo_1 + flujo_2 + flujo_3
```

```python
print("Total de flujos recibidos" , total_flujos)
```

```python
print(type(total_flujos))
```

# Comparaciones

```python
proyecto_rentable = total_flujos > inversion_inicial
```

```python
print("El proyecto recupera la inversion?" , proyecto_rentable)
```

```python
print(type(proyecto_rentable))
```
